Data is rendered onto a Figure. Each figure has one or more Axes objects ( not to be confused with Axis). It is the plot method on the Axes object that we use to a series. 

### Figure Introduction
The following basic sample creates a figure with one Axes object and draws a line series on the Axes.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots()

x_vals = [1,2,3,4]
y_vals = [2,4,6,8]

ax.plot(x_vals, y_vals)
```

### Axes 
Each Axes object is an area onto which we can plot data. If we are plotting in 2D each Axes object will have two Axis objects. The following code adds four Axes objects to a single figure

```python
fig, axes = plt.subplots(2,2)

noise = np.random.standard_normal(100)
walk = noise.cumsum()

axes[0,0].plot(noise)
axes[0,1].plot(walk)
axes[1,0].plot(noise)
axes[1,1].plot(walk)
```

### Configuring the Figure and Axes objects 

```python
x_values = np.linspace(0, 2, 100)  # Sample data.
fig, ax = plt.subplots()

# Plot a line series of linear values and give the series a lable of y=x



# Plot some data on the single Axes object. Note we need to add a legend to see the series lable "y=x". 
ax.plot(x_values, x_values, label='y=x')  

# Now a few more settings on the series. We set the color, linewidth, linestyle. We also use latex in the
# series label.
ax.plot(x_values, x_values**2, label=r'$y=x^2$', color='green', linewidth=3,linestyle='--') 
ax.legend()

# Label the x Axis and the y Axis. Note we optionally can set fontsize and color
ax.set_xlabel('x') 
ax.set_ylabel('y', color='purple', fontsize=18) 

# Set the title of the Axes object
ax.set_title('Polynomials')


# Annotate a point
ax.annotate('some point', xy=(1, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Add a grid to the Axes object
ax.grid(True)

#ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
#ax.plot(x, x**3, label='cubic')  # ... and some more.

#ax.set_title("Simple Plot")  # Add a title to the axes.


# Set the background color
#ax.set_facecolor('xkcd:light grey')
```

## Histograms

```python

# Generate 10000 standard normally distributed random numbers
normal_noise = np.random.randn(10000)

# Scale and shift the standard normal distribution to get a distribution with mean 100 and standard deviation 25
mu, sigma = 100, 25
values = (normal_noise * sigma) + mu

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.text(15, .0125, r'$\mu=100,\ \sigma=25$')
ax.set_xlabel('Stock Price')
ax.set_ylabel('Probability')
n, bins, patches = ax.hist(values, 100, density=True, facecolor='C0', alpha=0.75)
```
