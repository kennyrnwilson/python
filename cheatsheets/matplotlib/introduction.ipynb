# Introduction
### Hello Plot
The following code generates some noise and plots it.

```python
import matplotlib.pyplot as plt
import numpy as np

noise = np.random.standard_normal(100)
plt.plot(noise)
```

## Figure
Matplotlib plots live inside a figure. The following code creates figure with 2 subplots arranged in a grid with two rows and one column. Calling add_subplot creates a subplot and returns an AxesSubplot object. In general it is beter to use the axis methods rather the top level plt.plot

```python
fig = plt.figure()

sub1 = fig.add_subplot(2,1,1)
sub2 = fig.add_subplot(2,1,2)

walk1 = np.random.standard_normal(100).cumsum()
walk2 = np.random.standard_normal(100).cumsum()

sub1.plot(walk1,color="black", linestyle="solid" )
sub2.plot(walk1,color="green", linestyle="solid" )
```

## Add a Histogram
Alpha sets the transparency. I use 0.6

```python
fig = plt.figure()

sub1 = fig.add_subplot(2,2,1)
sub2 = fig.add_subplot(2,2,2)

noise1 = np.random.standard_normal(100)
noise2 = np.random.standard_normal(100)

sub1.hist(noise1, bins=10, color="green", alpha=0.6)
sub2.hist(noise2, bins=10, color="blue", alpha=0.6)
```

## Convenience Methods for creating a figure and axes

```python
col_count = 3; row_count = 2

figure, axes = plt.subplots(row_count,col_count)
for row_idx in range(row_count):
    for col_idx in range(col_count):
        axes[row_idx, col_idx].plot(np.random.standard_normal(100).cumsum(),color="black", linestyle="solid")
```

### Controlling Space between the axes
We can use the figure's subplots_adjust method to control space between the subplots

```python
col_count = 3; row_count = 2

figure, axes = plt.subplots(row_count,col_count)
for row_idx in range(row_count):
    for col_idx in range(col_count):
        axes[row_idx, col_idx].plot(np.random.standard_normal(100).cumsum(),color="black", linestyle="solid")
        
figure.subplots_adjust(wspace=0, hspace=0)
```

### Specifying Explicit X Values

```python
fig = plt.figure()

sub1 = fig.add_subplot(1,1,1)

xvals = np.arange(10,10000,100)
yvals = np.random.standard_normal(100).cumsum()

sub1.plot(xvals, yvals, color="black", linestyle="solid" )
```

### Colors as Hex
